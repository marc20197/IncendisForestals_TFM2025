```python
variables_escollides= [
    'Temperatura mitjana diària',
    'Temperatura màxima diària + hora',
    'Temperatura mínima diària + hora',
    'Humitat relativa mitjana diària',
    'Precipitació acumulada diària',
    'Ratxa màxima diària del vent 10 m + hora',
    'Direcció de la ratxa màx. diària de vent 10 m'
]
df_filtrat = df_sensecolumnes[df_sensecolumnes['NOM_VARIABLE'].isin(variables_escollides)]
```

```python
df_filtrat['VALOR'] = (
    df_filtrat['VALOR']
        .astype(str)                          # per si hi ha NaN
        .str.replace(',', '.', regex=False)   # coma → punt
)
```

```python
df_filtrat['VALOR'] = pd.to_numeric(df_filtrat['VALOR'])
```

```python
df_filtrat['VALOR'].dtype
```

```python
df_net = (
    df_filtrat
    .pivot_table(
        index=['CODI_ESTACIO', 'NOM_ESTACIO', 'DATA_LECTURA'],
        columns='NOM_VARIABLE',
        values='VALOR',
        aggfunc='first'
    )
    .reset_index()
)
df_net.index.name = None    

```

```python
df_net['DATA_LECTURA'] = pd.to_datetime(df_net['DATA_LECTURA'],
                                        dayfirst=True)
```

```python
df_net = df_net.sort_values('DATA_LECTURA').reset_index(drop=True)
```

<b>L'ÚLTIM QUE FARÉ SERÀ CANVIAR ELS NOMS DE LES COLUMNES PER FER-LOS MÉS SIMPLES</b>

```python
df_net = df_net.rename(columns={
    'Direcció de la ratxa màx. diària de vent 10 m': 'dir_ratxa_max_vent10',
    'Humitat relativa mitjana diària': 'hr_mitjana',
    'Humitat relativa mitjana diària': 'hr_mitjana',
    'Precipitació acumulada diària': 'prec_diaria',
    'Ratxa màxima diària del vent 10 m + hora': 'ratxa_max_vent10',
    'Temperatura mitjana diària': 't_mitjana',
    'Temperatura màxima diària + hora': 't_max',
    'Temperatura mínima diària + hora': 't_min'
})
```

```python
mask = df_net['DATA_LECTURA'].dt.year.between(2005, 2024)
df_estacions_2005_2024 = df_net.loc[mask].copy()
```

Ara el que farem serà afegir-hi les Columnes <b>Comarca</b> i <b>Codi_Comarca</b>. Ho traurem del fitxer de les metadades de les Estacions meteorològiques, on diu la Comarca a la que pertany cada Estació. AQUESTA INFORMACIÓ ES FARÀ SERVIR DESPRÉS PER UN MAPA COROPLÈCTIC DE LA METEOROLOGIA DE LES COMARQUES.

```python
import pandas as pd

metadades_estacions = pd.read_csv("Data/Meteorologia/Metadades_estacions_meteorològiques_automàtiques_20251130.csv")
```

```python
metadades_estacions_comarca = metadades_estacions[['CODI_ESTACIO', 'CODI_COMARCA', 'NOM_COMARCA']].drop_duplicates()
metadades_estacions_comarca
```

```python
df_final = df_estacions_2005_2024.merge(
    metadades_estacions_comarca,
    left_on='CODI_ESTACIO',   # al dataframe
    right_on='CODI_ESTACIO',  # al fitxer de metadades
    how='left'
)
```

```python
df_final
```

<b>Ja tenim el dataset arreglat.</b>
Finalment, l'escriurem en un fitxer csv. 

```python
df_final.to_csv("Meteo_Estacions.csv", index=False)
```

```python
for any in range(2005, 2025):
    df_any = df_final[df_final['DATA_LECTURA'].dt.year == any].copy()
    nom_fitxer = f"Meteo_Estacions_{any}.csv"
    df_any.to_csv(nom_fitxer, index=False)
    print(f"Guardat: {nom_fitxer} ({len(df_any)} files)")
```
